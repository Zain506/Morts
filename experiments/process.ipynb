1. Pull in image code: Use Chrome extension with browser cache. LIMIT NUMBER OF PULLS

```python
import requests
from PIL import Image
from io import BytesIO

# URL of the image
url = "https://media.licdn.com/dms/image/v2/D4E03AQEXpVSwl0ZOXw/profile-displayphoto-shrink_200_200/profile-displayphoto-shrink_200_200/0/1724105500186?e=1738800000&v=beta&t=1IBlD5wa98XaJjMyj--XylE4od6eCcfq-q7AapPggNY"

# Download the image
response = requests.get(url)

# Open the image using PIL
img = Image.open(BytesIO(response.content))

# Now you can work with the image, for example:
# Save it locally
img.save("linkedin_profile_picture.jpg")

# Or perform operations on it
# img = img.resize((100, 100))  # Resize the image
# img = img.convert('L')  # Convert to grayscale
```

2. Process image code

```python
from transformers import ViTFeatureExtractor, ViTModel
from PIL import Image
feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224-in21k')
model = ViTModel.from_pretrained('google/vit-base-patch16-224-in21k')
```

```python
inputs = feature_extractor(images=img, return_tensors="pt")
```

```python
outputs = model(**inputs)
last_hidden_state = outputs.last_hidden_state  # This contains the feature vectors
# To get a single feature vector for classification, you can use the [CLS] token's output:
cls_feature_vector = last_hidden_state[0, 0]  # The first element corresponds to [CLS] token
print(cls_feature_vector.shape)  # Should be (768,) for ViT base model
```

```python
vector = cls_feature_vector.detach().numpy()
print(vector)
```
